# problem1

choosing the average cosmic-ray background in 1 day (X) be the value of 5
then chossing the average number of gamma0rays emitted (Y) be the value of 8,
that is X = 5, Y = 8

```python
X = 5
Y = 8
#assume the probability distribution as a Poisson distribution
dist = stats.poisson(X)
x = np.arange(0, 15)
#plt(x, dist.pmf(x))

plt.vlines(x, 0, dist.pmf(x))
plt.ylim(0)

#complete all elements of the graph
plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of cosmic rays")
plt.ylabel("probability mass")
```

Then, when we integrate (sum) for more days, we need to add up the probability mass function for more times
Similar with homework 2 problem 5, we know that the sum and the average tend to the Guassian distribution.

```python
pmf2 = np.convolve(dist.pmf(x), dist.pmf(x)) #sum for two days
plt.vlines(range(len(pmf2)), 0, pmf2)
plt.yscale('log')
plt.xlim(0, 30)
plt.ylim(10**(-21), 1)

plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of sum of cosmic rays")
plt.ylabel("probability mass")
```

```python
# then we sum for five days
pmfNew = dist.pmf(x)
for i in range(1, 5):
    pmfNew = np.convolve(pmfNew, dist.pmf(x))
plt.vlines(range(len(pmfNew)), 0, pmfNew)
plt.yscale('log')
plt.xlim(0, 80)
plt.ylim(10**(-21), 1)

plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of sum of cosmic rays")
plt.ylabel("probability mass")
```

```python
# then we sum for twenty days
pmfN = dist.pmf(x)
for i in range(1, 20):
    pmfN = np.convolve(pmfN, dist.pmf(x))
plt.vlines(range(len(pmfN)), 0, pmfN)
plt.yscale('log')
plt.xlim(0, 250)
plt.ylim(10**(-21), 1)

plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of sum of cosmic rays")
plt.ylabel("probability mass")
```

From the graph, we can find out that the summed probability distribution is still Poisson after five days.

c) Then we come to the average part, we will try the average of ten days, and then the average of twenty days.

```python
# the average of ten days
pmfNew2 = dist.pmf(x)
for i in range(10):
    pmfNew2 = np.convolve(pmfNew2, dist.pmf(x))
plt.vlines(np.arange(len(pmfNew2)) / 11, 0, pmfNew2)
plt.xlim(0)
#plt.ylim(0)
plt.yscale('log')
plt.ylim(1e-10, 1);
plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of average of cosmic rays")
plt.ylabel("probability mass")
```

```python
# the average of twenty days
pmfNew3 = dist.pmf(x)
for i in range(20):
    pmfNew3 = np.convolve(pmfNew3, dist.pmf(x))
plt.vlines(np.arange(len(pmfNew3)) / 21, 0, pmfNew3)
plt.xlim(0)
plt.ylim(0)
plt.yscale('log')
plt.ylim(1e-10, 1);
plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of average of cosmic rays")
plt.ylabel("probability mass")
```

```python
# the average of hundred days
pmfNew4 = dist.pmf(x)
for i in range(100):
    pmfNew4 = np.convolve(pmfNew4, dist.pmf(x))
plt.vlines(np.arange(len(pmfNew4)) / 101, 0, pmfNew4)
plt.xlim(0)
plt.ylim(0)
plt.yscale('log')
plt.ylim(1e-10, 1);
plt.title("Poisson distribution of cosmic-ray")
plt.xlabel("The number of average of cosmic rays")
plt.ylabel("probability mass")
```

With the increasing of the number os days, the distribution is much closer to the Guassian distribution. This tendency also shows in the part b). 
From the central limit theorem, we know that when we sum up the random independent variables, then their corresponding normalized sum will tend to the Guassian distribution. This situation happens even when the variables are not initially normal distributed.

# problem2

Choosing the Rayleigh distribution, where the scale parameter is sigma. Here, we can choose the sigma as the value of 0.8.

```python
print('\u03C3 = 0.8')
```

```python
dist2 = stats.rayleigh(scale = 0.8) #create a new Rayleigh distribution
x = np.linspace (0,10,1000) #range of x values
pdf =  dist2.pdf(x)
plt.plot(x, pdf)
plt.ylim(0)
plt.title("Rayleigh distribution's pdf")
plt.xlabel("x values")
plt.ylabel("probability density")
```

For the different observing intervals, we first try 10, 20, 30 intervals.
For the convenience, we can plot the graph in one 

```python
pdfNew = pdf
for j in range(3):
    for i in range(10):
        pdfNew = np.convolve(pdfNew, pdf)
    plt.plot(np.linspace(0, 10, len(pdfNew)), pdfNew)
plt.yscale("log")

# the green line is 30 intervals, the orange line is 20 intervals, and the blue line is 10 intervals
```

```python
pdfNew = pdf
for j in range(2):
    for i in range(50):
        pdfNew = np.convolve(pdfNew, pdf)
    plt.plot(np.linspace(0, 10, len(pdfNew)), pdfNew)

plt.yscale("log")
#the blue line is 50 intervals, and the orange line is 100 intervals
```

We find that with the increasing of the observing intervals, shapes become much thinner and taller. Meanwhile, the extent of the skew is much smaller, and the distribution tends to be Guassian distribution.

```python
#200
new_pdf = pdf/100 #ajusted by dividing 100 to make the plot is able to plot
for i in range(200):
    new_pdf = np.convolve(new_pdf, pdf/100)
plt.semilogy(np.linspace(0, 10, len(new_pdf)), new_pdf)
plt.ylim(1e-50)
```

With several trials, I find that it approaches the Guassian distribution after about 200 intervals.

# problem 3

choosing x, the width of the background Gaussian distribution 5, 
and y, the signal of strengh, be 35.3

```python
dist = stats.norm(scale=5) #where the width of the background Guassian is 5
x = np.linspace(0, 30, 1000) #x-values
plt.plot(x, dist.pdf(x))
plt.xlim(0)
plt.ylim(0)
plt.title("background noise's pdf for a single pixel")
plt.xlabel("brightness unit")
plt.ylabel("probability density")
```

Here, we want to explore the probability that the signal-free background produced the measurement that is signal-like or more signal-like than 35.3 units? Then, we convert this problem into the integral, it should be the pdf from 35.3 to the positive infinite.

```python
prob = 1-dist.cdf(35.3)
sigma=stats.norm.ppf(1 - prob)
```

```python
print(prob) #the probability 
print(sigma) #converted to the sigma, is about 7.06
```

We find the significance here is about 7.06 sigma, is large than 5 sigma. Hence, the null hypothesis can be rejected.
We can claim the discovery.

# version two

This time we consider about 10000 pixels. Similarly, we need combine all 10000 pixels' pdf. Here, we can easily multiply 10000 to get the new pdf.

```python
plt.plot(x, 10000 * dist.pdf(x))
plt.xlim(0, 30)
plt.ylim(0);

plt.title("background noise's pdf for 10000 pixels")
plt.xlabel("brightness unit")
plt.ylabel("probability density")
```

From the graph, we can find out that the probability density is also multiple 10000 times.
Here, we can explore the same question as version 1. 
We want to explore the probability that the signal-free background produced the measurement that is signal-like or more signal-like than 35.3 units of 10000 pixels? Then, we convert this problem into the integral, it should be the 10000*pdf from 35.3 to the positive infinite.

```python
prob2 = (1-dist.cdf(35.3))*10000
stats.norm.ppf(1 - prob2)
```

here this sigma is still larger than 5 sigma, so our discovery still can be made.
